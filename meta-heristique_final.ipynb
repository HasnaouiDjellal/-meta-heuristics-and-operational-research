```python
def calc_som_poids(v,vp):   #somme des poids
    sam = 0
    for j in range(0,len(v)):
        sam =sam + v[j]*vp[j]
    return sam    
```

```python
def calc_som_porfait(v,vpr):  #somme des porfait
    sam = 0
    for j in range(0,len(v)):
        sam =sam + v[j]*vpr[j]
    return sam 
```

```python
from random import randint  #voisinage 1
def voisinage1(s1):
    for i in range(0,5):
        r = randint(0,9)
        if (s1[r] == 0):
            s1[r] = 1 
        else:
            s1[r] = 0
    return s1
```

```python
from random import randint #voisinage 2
def voisinage2(s2):
    
    r1 = randint(0,9)
    r2 = randint(0,9)
    if (r1 != r2):
        per = s2[r1]
        s2[r1] = s2[r2]
        s2[r2] = per
    return s2
```

```python
def voisinage3(s3):
    for i in range(0,len(s3)): #voisinage 3
        s3[i] = randint(0,1)
    return s3
```

```python
def voisinage4(s4):
    ranb = randint(0,3)
    rand = ranb + 2
    while ranb != rand : #voisinage 4
        per = s4[ranb]
        s4[ranb] = s4[rand]
        s4[rand] = per 
        ranb = ranb + 1
        rand = rand - 1
    return s4    
   
```

```python
def glouton(profit, poids, capacity):
    # index = [0, 1, 2, ..., n - 1] for n items
    index = list(range(len(profit)))
    # contains ratios of values to weight
    ratio = [v/w for v, w in zip(profit, poids)]
    # index is sorted according to value-to-weight ratio in decreasing order
    index.sort(key=lambda i: ratio[i], reverse=True)
 
    poids_total = 0
    profit_total = 0
    s = [0]*len(profit)
    for i in index:
        if poids[i] <= capacity:
            s[i] = 1
            profit_total += profit[i]
            poids_total += poids[i]
            capacity -= poids[i]
    print("La solution optimal est : ",s)
    print("Le poids total est : ",poids_total)
    print("Le profit total est : ", profit_total) 
 
    return s

```
